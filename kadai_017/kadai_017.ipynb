```python
class Human:
    def __init__(self, name, age):
        self.name = name
        self.age = age

    def check_adult(self):
        if self.age >= 20:
            print(self.name, "は大人です")
        else:
            print(self.name, "は大人ではありません")


# Humanクラスのインスタンスを複数生成
humans = [
    Human("貴志", 25),
    Human("孝子", 18),
    Human("一郎", 20)
]

# リストの要素数分だけcheck_adultメソッドを呼び出す
for human in humans:
    human.check_adult()
```
